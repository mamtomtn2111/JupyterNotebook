```python
#Nguyn V Hi _ CNTT K18 CLC
  
import cv2         
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(10, 5))
  
image1 = cv2.imread('Coin.jpg')

image1 = cv2.resize(image1, (1280, 1200))      
#Convert nh v a cp xm

img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
# dng thut ton otsu km
# phn on nh phn nh

ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY + 
                                            cv2.THRESH_OTSU)     
#Convert li mu chun cho nh
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image_output = cv2.cvtColor(thresh1, cv2.COLOR_BGR2RGB)

# Dng th vin matplotlib  v ra hnh minh ha       
plt.subplot(121)
plt.imshow(image1)
plt.subplot(122)
plt.imshow(image_output)  
```
