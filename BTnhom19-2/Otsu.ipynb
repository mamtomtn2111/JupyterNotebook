```python
#Nguyễn Vũ Hải _ CNTT K18 CLC
  
import cv2         
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(10, 5))
  
image1 = cv2.imread('Coin.jpg')

image1 = cv2.resize(image1, (1280, 1200))      
#Convert ảnh về đa cấp xám

img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
# dùng thuật toán otsu kèm
# phân đoạn nhị phân ảnh

ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY + 
                                            cv2.THRESH_OTSU)     
#Convert lại màu chuẩn cho ảnh
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image_output = cv2.cvtColor(thresh1, cv2.COLOR_BGR2RGB)

# Dùng thư viện matplotlib để vẽ ra hình minh họa       
plt.subplot(121)
plt.imshow(image1)
plt.subplot(122)
plt.imshow(image_output)  
```
