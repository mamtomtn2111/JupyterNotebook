### Load ảnh cần phân cụm màu

```python
image = cv2.imread("test.jpg")
plt.figure(figsize = (15,8))
plt.imshow(image)
image.shape
```

### Chuyển đổi từ RGB sang HSV

```python
hsv_img = cv2.cvtColor(image ,cv2.COLOR_BGR2RGB)
#  reshape thành 3 cột 
vectorized = hsv_img.reshape((-1,3))
# convert sang float32
vectorized = np.float32(vectorized)
```

Xác định tiêu chí, số lượng cụm (K) và áp dụng K-means

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
```

TERM_CRITERIA_EPS sẽ dừng thuật toán nếu đạt đến độ chính xác chỉ định, giá trị ở đây là 1.0
TERM_CRITERIA_MAX_ITER sẽ dừng thuật toán sau số lần lặp được chỉ định, giá trị ở đây là 10

#### Giải thích ý nghĩa cv2.KMEANS_PP_CENTERS:
Với biến này xuất hiện trong hàm, đầu tiên opencv sẽ lặp lại cả hình ảnh để xác định tâm và sau đó mới bắt đầu phân cụm.

Ưu điểm:
Sẽ cho ra kết quả tối ưu và đều đặn trong mỗi lần chạy nếu như cùng 1 ảnh input.
Nhược điểm:
Tốn thời gian xử lý hơn cho việc xác định tâm.


```python
K = 4  # số cụm xác định trước
attempts = 10

ret,label,center = cv2.kmeans(vectorized,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)

# đổi sang dạng 8 bit
center = np.uint8(center)

# đổi sang mảng một chiều
res = center[label.flatten()]

# reshape về mảng như mảng ảnh ban đầu
result_image = res.reshape((hsv_img.shape))

#Resize đồ thị hiển thị ra
plt.figure(figsize=(15,15))

#Hiển thị ảnh đã qua xử lý
plt.subplot(1,2,1)
plt.imshow(hsv_img)
plt.title('Ảnh gốc')
plt.xticks([])  #tắt xticks
plt.yticks([])  #tắt yticks

plt.subplot(1,2,2)
plt.imshow(result_image)
plt.title('Ảnh đã phân cụm khi có K = %i' % K)
plt.xticks([])
plt.yticks([])

plt.show()
```
