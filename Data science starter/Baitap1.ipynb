2.6.8 sắp xếp lại dữ liệu
Kiểu function này là dùng để hiển thị các loại dữ liệu hay lặp lại theo từng cột, hàng, để hiển thị specific dạng kiểu table\

2.6.10 Plotting

```python
#Dùng biểu đồ countplot và kỹ thuật bbinning chia thành 16 khoảng điểm và đếm các khoảng điểm đó
nbins = 15
arrDiem = np.array(K18Mark4['Điểm hệ 4'])
bins = np.linspace(arrDiem.min()-0.0001, arrDiem.max(), nbins + 1)
print(f"Number of interval {nbins} bins is calculated:")
display(bins)
K18Mark4["binned"] = pd.cut(arrDiem, bins)
display(K18Mark4.head())
print(K18Mark4['binned'].value_counts())
```

```python
figure(figsize = (25,10))
plt.title("Điểm khoa CNTT ICTU KÌ 2 NĂM HỌC 2020 - 2021", size = 15)
ax = sns.countplot(x = "binned", data = K18Mark4, alpha = 0.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 40, ha = "right")
plt.xlabel("Khoảng điểm (từ 0 - 4)", size = 15)
plt.ylabel("Số sinh viên", size = 15)
```

```python
totalSum = K18Mark4["binned"].value_counts()
print(totalSum
totalSum.plot(kind = 'bar', style = 'b', alpha = 0.4, title = 'Tổng số điểm của học sinh')
```
