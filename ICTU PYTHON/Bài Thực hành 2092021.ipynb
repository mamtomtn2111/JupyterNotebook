```python
import math
print("Solve quadractic equation ax^2 + bx + c = 0")
a = float(input("Input a number: "))
b = float(input("Input b number: "))
c = float(input("input c number: "))

if(a == 0 and b == 0 and c == 0):
    print("The equation has uncountable solution !")
elif(a == 0 and b == 0 and c != 0):
    print("The equation has no solution !")
elif(a == 0):
    solution = -c / b
    print("The equation has only one solution x = ",solution)
else:
    discriminant = b*b - 4*a*c
    if(discriminant < 0):
        print("The equation has no solution !")
    elif(discriminant == 0):
        solution = - b / (2*a)
        print(f"The equation has two same solutions x1 = x2 = {solution}")
    elif(discriminant > 0):
        solution1 = (-b + math.sqrt(discriminant)) / (2*a)
        solution2 = (-b - math.sqrt(discriminant)) / (2*a)
        print(f"The equation has two solutions x1 = {solution1} , x2 = {solution2}")
    else:
        print("Exceptions !")
```

```python
# Quý Trần Quang14:28
# Gợi ý: D = a1b2-a2b1
# Dx= c1b2 – c2b1
# Dy = a1c2 – a2c1
# 1.	Nếu D = 0 và Dx =0 và Dy= 0 thì: Phương trình vô số nghiêm
# 2.	Nếu D = 0 (Dx#0 hoặc Dy #0 ) thì: Phương trình vô nghiệm
# 3.	Nếu D #0 thì hệ phương trình có nghiệm: x = Dx/D và y = Dy/D

print("Giải hệ pt 2 ẩn")
print("a1x + b1y = c1")
print("a2x + b2y = c2")

a1 = int(input("Nhập vào số a1: "))
b1 = int(input("Nhập vào số b1: "))
c1 = int(input("Nhập vào số c1: "))
a2 = int(input("Nhập vào số a2: "))
b2 = int(input("Nhập vào số b2: "))
c2 = int(input("Nhập vào số c2: "))

D = a1 * b2 - a2 * b1
Dx= c1 * b2 - c2 * b1
Dy = a1 * c2 - a2 * c1

if(D == 0 and Dx == 0 and Dy == 0):
    print("Hệ PT vô số nghiệm")
elif(D == 0 and Dx != 0 and Dy != 0):
    print("Hệ PT vô nghiệm")
elif(D != 0):
    x = Dx/D
    y = Dy/D
    print(f"PT có 2 nghiệm x = {x}, y = {y}")
```

```python
#Bài 5: Viết chương trình nhập vào từ bàn phím kiểm tra một số có phải số chính phương hay không?
from math import *
def CheckChinhPhuong(n):
    sqrtN = int(sqrt(n))
    if(sqrtN * sqrtN == n):
        return 1
    else:
        return 0
    
n = int(input("Mời nhập vào số n: "))
if(CheckChinhPhuong(n)):
    print(f"Số {n} là số chính phương")
else:
    print(f"Số {n} không phải là số chính phương")
```

```python
def CheckSoHoanHao(n):
    sum=0
    for i in range(1,n):
        if n % i==0:
            sum+= i
    if sum == n:
        return 1
    else:
        return 0
    
n = int(input("Nhap n : "))
for i in range(1,n):
    if(CheckSoHoanHao(i)):
        print(i)
```

```python
import math
def snt(n):
    if n < 2:
        return False
    for i in range(2,int(math.sqrt(n))+1):
        if(n % i == 0):
            return False
    else:
        return True
n = int(input("Nhap n : "))
for i in range(1,n+1):
    if(snt(i)):
        print(i)
```

```python
def fibonacci(n):
    if(n == 1 or n == 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
    
n = int(input("Nhập vào số n:"))
print("n số đầu tiên của dãy số fibonacci: ");
for i in range(1,n+1):
    print(fibonacci(i))
```
