```python
#THUẬT TOÁN SẮP XẾP TRỘN: MERGE SORT 

def Merge_sort(arr):
    start = 0
    end = len(arr) - 1
    
    temp = arr.copy()

    divide = 1
    while divide <= len(arr):
        for i in range(start, end, 2 * divide):
            From = i #0
            mid = i + divide - 1 #0
            to = min(i + 2 * divide - 1, end) #1
            merge(arr, temp, From, mid, to)
            
        divide *= 2
            
def merge(arr, temp, From, mid, to):
    FromArr = From #0
    FromTemp = From #0
    #to = 1
    #mid = 0
    Mid1 = mid + 1 # 1

    #Các phần tử sẽ được sắp xếp và ghép với nhau đối với to > 1
    while FromArr <= mid and Mid1 <= to:
        if arr[FromArr] < arr[Mid1]:
            temp[FromTemp] = arr[FromArr]
            FromArr += 1
        else:
            temp[FromTemp] = arr[Mid1]
            Mid1 += 1
        FromTemp += 1
        
    #Kiểm tra xem còn phần tử sót lại, và điền vào mảng tạm, để thế lên mảng chính
    while FromArr < len(arr) and FromArr <= mid:
        temp[FromTemp] = arr[FromArr]
        FromArr += 1
        FromTemp += 1
        
    #Sao chép lại mảng chính
    for i in range(From, to + 1):
        arr[i] = temp[i]
        
if __name__ == '__main__':
    
    arr = [9, 4, 4.5, 10, -5, 1, 5]
    print(f"Dãy được cho: {arr}")
    Merge_sort(arr)
    print(f"Dãy được sắp xếp theo mergeSort: {arr}")
```

```python
from tkinter import *
from tkinter import ttk
import numpy as np
import time
```

```python
def show(n: int, data: list, colours: list):
 
    canvas.delete('all')
 
    width = 1560/(3*n-1)
 
    gap = width/2
 
    for i in range(n):
 
        # this function will display an array of "bar"
        canvas.create_rectangle(7+i*width+i*gap, 0, 7 +
                                (i+1)*width+i*gap, data[i],
                                fill=colours[i])
     
    win.update_idletasks()
 
 
def shuffle():
    np.random.shuffle(arr)
    show(N, arr, color)
 
 
def mergesort(arr, left, right):
    if left < right:
        m = (left+right)//2
        mergesort(arr, left, m)
        mergesort(arr, m+1, right)
 
        j = m+1
        if arr[m] <= arr[m+1]:
            return

        while left <= m and j <= right:
            show(N, arr, ['blue' if x == left or x ==
                          j else 'grey' for x in range(N)])
            time.sleep(1/speed)
            if arr[left] <= arr[j]:
                left += 1
            else:
                show(N, arr, ['red' if x == left or x ==
                              j else 'grey' for x in range(N)])
                 
                # array of colours where only the focused bars
                # are displayed red since left >arr[j]
                time.sleep(1/speed)
                temp = arr[j]
                 
                # storing the smaller element in temp variable
                i = j
                while i != left:
                    arr[i] = arr[i-1]
                    show(N, arr, ['red' if x == i or x ==
                                  j else 'grey' for x in range(N)])
                    time.sleep(1/speed)
                    i -= 1
                 
                arr[left] = temp
 
                show(N, arr, ['green' if x == left or x ==
                              j else 'grey' for x in range(N)])
                time.sleep(1/speed)
                left += 1
                m += 1
                j += 1

def start():
    mergesort(arr, 0, N-1)
    show(N, arr, ['purple' for _ in range(N)])
 
 
win = Tk()
 
N = 50  
speed = 10
 
arr = np.linspace(10, 390, N, dtype=np.uint16)
 
color = ['grey' for _ in range(N)]
 
ttk.Label(win, text='Merge Sort visualizer - Nhóm 6').pack()
canvas = Canvas(win, width=800, height=400, bg='white')
canvas.pack()
 
ttk.Button(win, text='Bắt đầu', command=start).pack(
    side='right', padx=5, pady=5)
 
ttk.Button(win, text='Trộn mảng', command=shuffle).pack(side='right')
shuffle()
show(N, arr, color)
 
win.mainloop()
```
