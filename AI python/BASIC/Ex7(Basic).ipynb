```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.gridspec as gridspec
sns.set()
```

```python
flights = sns.load_dataset('flights')
display(flights.head())
print(flights.shape)
print(f"Min Year: {flights.year.min()} and Max Year: {flights.year.max()}")
print(f"Min Passenger: {flights.passengers.min()} and Max Passenger: {flights.passengers.max()}")
print(f'Number of Unique Months : {list(flights.month.unique())}')
print("****Frequency****")
print(f"<<<Count of months>>> \n{flights.month.value_counts()}")
print(f"<<<Count of year>>> \n{flights.year.value_counts()}")
```

```python
tips = sns.load_dataset('tips')
display(tips.head())
print(tips.shape)
print(f"Min total_bill: {tips.total_bill.min()} and Max total_bill: {tips.total_bill.max()}")
print(f"Min tip: {tips.tip.min()} and Max tip: {tips.tip.max()}")
print(f'Number of Unique sex : {list(tips.sex.unique())}')
print(f'Number of Unique smoker : {list(tips.smoker.unique())}')
print(f'Number of Unique day : {list(tips.day.unique())}')
print(f'Number of Unique time : {list(tips.time.unique())}')
print(f'Number of Unique size : {tips["size"].unique()}')
print("****Frequency****")
print(f"<<<Count of sex>>> \n{tips.sex.value_counts()}")
print(f"<<<Count of smoker>>> \n{tips.smoker.value_counts()}")
print(f"<<<Count of day>>> \n{tips.day.value_counts()}")
print(f"<<<Count of time>>> \n{tips.time.value_counts()}")
print(f'<<<Count of size>>> \n{tips["size"].value_counts()}')
```

```python
# Question 1: Visualization the column passengers in flights dataset with the following requirements:
# A. Create equally interval bins by min and max of passengers 
# B. Plot the histogram of each bins using hist and countplot
# C. Show the percent on the top of column in each interval

nbins = 15
arr = np.array(flights["passengers"])
bins = np.linspace(arr.min(), arr.max(), nbins + 1) #nbins +1 Chia thành 15 khoảng + 1
print(f"Number of interval {nbins} bins is calculated : ")
display(bins)
flights['binned'] = pd.cut(arr, bins)
display(flights.head())
print(flights["binned"].value_counts())
```

```python
figure(figsize=(25, 5))

# Using Histogram
plt.subplot(121)
plt.title("Histogram of Passengers", size = 15)
plt.xlabel("Bin of Passengers", size = 15)
plt.ylabel("Frequency of each bins", size = 15)
density , bins, _ = plt.hist(flights["passengers"], bins = 15, alpha=0.5, color = "g")
# count, _ = np.histogram(flights["passengers"], bins)
for x,y in zip(bins, density):
    if int(y) != 0:
        plt.text(x, y+0.05, int(y), fontsize=15) # x,y,str 
        
# Using CountPlot
plt.subplot(122)
plt.title("Count Plot of Passengers by Bins", size = 15)
ax = sns.countplot(x="binned", data=flights, alpha=0.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.xlabel("Bin of Passengers", size = 15)
plt.ylabel("Count of each bins", size = 15)

for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(flights))
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() 
    ax.annotate(percentage, (x, y),ha='center',size = 10)

plt.show()
```

```python
# Question 2: Visualization the column total_bill in tips dataset with the following requirements:
# A. Comparision of total_bill of Male and Female
# B. Comparison of total_bill and day by smoker 
# C. Distribution of tips and total_bill by time
```

```python
tips_male = tips[tips["sex"] == "Male"]
display(tips_male.head())
tips_female = tips[tips["sex"] == "Female"]
display(tips_female.head())
```

```python
figure(figsize=(15,8))
sns.histplot(tips_male['total_bill'],color = "r",kde = True)
plt.show()
figure(figsize=(15,8))
sns.histplot(tips_female['total_bill'], color = "b", kde = True)
plt.show()
```

```python
fig, axes  = plt.subplots(nrows=2, ncols=2)
fig.set_figwidth(15)
fig.set_figheight(10)
fig.suptitle("Bi-Variate Analysis")

sns.histplot(tips_male['total_bill'], ax=axes[0, 0], kde = True, color = "r")
sns.histplot(tips_female['total_bill'], ax=axes[0, 0], kde = True, color = "b"); 
axes[0, 0].set_title('Male and Female Total Bill')

ax_colspan = plt.subplot2grid((2, 2), (0, 1), rowspan=2)
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3", ax=ax_colspan)
ax_colspan.set_title('Box plot of Day and Total_Bill')

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", ax=axes[1, 0])
axes[1, 0].set_title('Distribution tip and total_bill')

plt.show()
```

```python
# Question 3 : Visualization the number of passenger on year and month by
# A. Compare number of passengers by month year
# B. Using Line, Headmap, Barplot and Bubble Plot
```

```python
flights_wide = flights.pivot("year", "month", "passengers")
flights_wide.head()
```

```python
# Setup the subplot2grid Layout
fig = plt.figure(figsize=(15, 12))
ax1 = plt.subplot2grid((3,5), (0,0), colspan = 3, rowspan = 2) 
ax2 = plt.subplot2grid((3,5), (0,3), colspan = 2, rowspan = 2)
ax3 = plt.subplot2grid((3,5), (2,0), colspan = 2)
ax4 = plt.subplot2grid((3,5), (2,2), colspan = 3)

# Time Series
sns.lineplot(data=flights, x="year", y="passengers", hue="month", ax = ax1)
sns.scatterplot(data=flights, x="year", y="passengers", hue="month",legend=False, ax = ax1)
ax1.set_title("Time Series")

# Scatterplot
#dùng pivot để tạo heatmap
sns.heatmap(flights_wide, annot=True, fmt="d", ax = ax2)
ax2.set_title("Heatmap")

# BarPlot
#Dấu gạch nhỏ thể hiện dao động
sns.barplot(x='year',y='passengers',data=flights, ax = ax3)
ax3.set_title("Barplot")

# Correlation
sns.scatterplot(data=flights, x="year", y="month", size="passengers", legend=False, sizes=(20, 2000), ax = ax4)

fig.tight_layout()
```
