```python
# Question 1: Input from Keyboard a list of Fruit and do 
# A. Remove the duplicate fruits 
# B. Convert to Array
# C. Print the number of Fruits 

#design solutions
#B1. Nhap danh sach list tu ban phim ==> input string ==> split thanh list
#b2. Xoa cac trai cay trung nhau ==> chuyen list sang set tro lai list
#B3 chuyen kieu tu list sang array ==> np.array(truyen list vao)
#B4. Xuat so luong cac trai cay co trong Array ==> len(truyen array vao)
#Implementation
#B1
print("#B1")
strInput = input('Please input a list of fruits separated by comma ",": ')
#strinput = Apple,Orange,Apple,Strawberry,banana
FruitList = strInput.split(",")
#B2
print("#B2")
FruitList = list(set(FruitList))
#print(FruitList) #Xuat thuong
display(FruitList) #Xuat theo format
#B3
print("#B3")
FruitArray = np.array(FruitList)
print(type(FruitArray))
display(FruitArray)
#B4
print("#B4")
print(len(FruitArray))
print(f'The number of fruits is {len(FruitList)} and the first is "{FruitList[0]}" ')
print(f"The number of fruits is {len(FruitArray)} and the last is \"{FruitArray[-1]}\" ")
```

```python
# Question 2: Input from Keyboard a list of Country and City and do 
# A. Create a Series from Country and City in which  Country is key and City is value 
# B. Allow user input a key and we print the corresponding values 

# Design solutions
#B1. Nhap vao danh sách các country và city: Nhập string ==> split ==> list
#B2. Tạo một series từ hai danh sách countrylist và citylist: pd.series(CityList, index = countrylist)
#B3/ Nhập một key là country từ bàn phím, và dùng series tìm kiếm city của country đó : input ==> Series[Country]

#implemantation
#B1
print("#B1")
strCountry = "USA,Viet Nam,Korea"
strCity = "New york,HCM City,Seoul"
CountryList = strCountry.split(",")
CityList = strCity.split(",")
display(CountryList)
display(CityList)

#B2
print("#B2")
CountrySeries = pd.Series(CityList, index = CountryList)
print(type(CountrySeries))
display(CountrySeries)
print(CountrySeries["Viet Nam"])

#B3
print("#B3")
inputCountry = input("Fill in country:")
print(CountrySeries[inputCountry])
print(CountrySeries.index)
print(CountrySeries.values)
if inputCountry in CountrySeries.index:
    print(CountrySeries[inputCountry])
else:
    print("Your index is not in the list")
```

```python
# Question 3: Input from Keyboard a list of Colors and do 
# A. Make sure that all the input color in lowercase  
# B. Remove the duplicate Colors and remove the color "Black" in list
# C. Convert to Array and Print the number of Colors 

#B1. Nhap mot chuoi mau : strColor
#B2. CHuyen chuoi mau na thanh chu roi tao danh sach mau: lower ==> split ==> colorlist
#B3. Xoa mau trung nhau trong danh sach: list(set(Truyen vao list))
#B3. Xoa mau den trong danh sach: list(set(truen vao danh sach) - set(["black"])) or np.remove()
#B4. Chuyen tu list sang array
colorInput = input('Input colors, each color is separated by comma ",": ')
#String = red,RED,red,Blue,bLUE,Green
colorInput = colorInput.lower()
print("Colors were input and made lowercase: ",colorInput)
ColorList = colorInput.split(",")
ColorList = list(set(ColorList))
print("Remove the duplicate colors: ",ColorList)
ColorList = set(ColorList) - set(["black"]) ; ColorList = list(ColorList) #Change data set to list
print('Remove the color "black": ',ColorList)
ColorArray = np.array(ColorList)
print(f"The number of color is: {len(ColorArray)}")
```

```python
# Question 4: Input from Keyboard a list of StudentName and Maths Marks and do 
# A. Create a Series from StudentName and Marks in which StudentName is key and Marks is value 
# B. Allow user input a name of student and we print the corresponding marks
# C. Print the average and std Marks of all students
#B1. Nhap tu ban phim chuoi studentNam va chuoi MathMarks sau do chuyen thanh list : input ==> split
#B2. Tao series voi key la StudentNameList va value la Marklist : pd.series(Truyen vao ds MarkList, index = dssv)
#B3. Nhap mot ten sinh vien va in ra diem: inputName ==> Series ==> series[inputName]
#B4. Tinh mea va std cua danh sach diem : math.mean va math.std tren Series.values
#Step A
import statistics
StudentName = "Dung,Hai,John,Peter,Jane,Doan"
Mark = "8.5,10,4.6,7.8,9.4,8"
StudentList = StudentName.split(",")
MarkList = Mark.split(",")
print(StudentList)
print(MarkList)
#MarkList = [float(i) for i in MarkList] #Convert all string into float or int in a list
MarkList = map(float, MarkList) #Convert all string into float or int in a list
StudentSeries = pd.Series(MarkList,index = StudentList)
#Step B
InputStudent = input("Please input a name of student and display the mark: ")
if InputStudent in StudentSeries.index:
    print(StudentSeries[InputStudent])
else:
    print("Student's name is not valid")
#Step C
print("Average of student's mark is: ",statistics.mean(StudentSeries.values))
print("Standard Deviation of student's mark is: ",statistics.stdev(StudentSeries.values))
```

```python
# Question 6: Setup a List, Series, and Array, Tuple, Dict from a list of random number and do 
# A. Input a number and add to all data strutures  
# B. Input a number and search if exists and print the position 
# C. Convert all data structures to List

import random
random.seed(12)
NumList = random.sample(range(10, 100), 5)

print("==" * 5 + "Convert to other data structures " + "==" * 5)
NumArray = np.array(NumList)
NumTuple = tuple(NumList)
NumDict = dict(zip(range(len(NumList)), NumList))
NumSeries = pd.Series(NumList)

print(f" List : {NumList} and Type : {type(NumList)} and NElement : {len(NumList)}")
print(f" Array : {NumArray} and Type : {type(NumArray)} and NElement : {len(NumList)}")
print(f" Tuple : {NumTuple} and Type : {type(NumTuple)} and NElement : {len(NumList)}")
print(f" Dictionary : {NumDict} and Type : {type(NumDict)} and NElement : {len(NumList)}")
print(f" Series : \n{NumSeries} and Type : {type(NumSeries)} and NElement : {len(NumList)}")

print("==" * 5 + "Input a Number and print Position in List " + "==" * 5)
inputNum = 90
IdxFound = NumList.index(inputNum) if inputNum in NumList else -1/2
print(f"Search the value {inputNum} in List : {NumList} at position : {IdxFound}")

print("==" * 5 + "Convert all data structures to List " + "==" * 5)
NumList1 = list(NumArray) 
NumList2 = list(NumTuple)
KeyValuesList3 = list(NumDict.items())
NumList3 = [value for key,value in KeyValuesList3]
KeyValuesList4 = list(NumSeries.items())
NumList4 = [value for key,value in KeyValuesList4]
print(NumList1, NumList2, NumList3, NumList4)
```

```python
# Special Remember CRUD on a List : Create, Read, Update, and Delete an elements
import random
print("==" * 5 + "CRUD on a List : Create, Read, Update, and Delete an elements" + "==" * 5)
random.seed(15)

NewNumList = random.sample(range(10, 100), 10)
print("Create a Random List :", NewNumList)
NewNumList1 = NewNumList[1:6]
print("Read Element from 1 to 5 :", NewNumList1)
NewNumList2 = NewNumList.copy()
NewNumList2[5] = 100
print("Update Element at position 5 :", NewNumList2)
NewNumList2.insert(8, 200)
print("Insert Element at position 8 :", NewNumList2)
NewNumList2.append(300)
print("Append Element at the last :", NewNumList2)
NewNumList3 = NewNumList2 + [400,500]
print("Concat another List to List :", NewNumList3)
NewNumList3.remove(400)
print("Delete an element from List :", NewNumList3)
NewNumList3.sort()
print("Sort a List :", NewNumList3)
NewNumList3.reverse()
print("Reverse a List :", NewNumList3)
NewNumList4 = [NewNumList3[index] for index in [1,3,5,7]] 
print("Get a List by index :", NewNumList4)
```

```python
# Question 7: Input a List of your Plan (For Example: Hiking at weekend, Swimming on Monday, etc) and do
# A. Input any hobby and find any plan having your hobby or not   
# B. Change that plan with another plan 
# C. Input a list of Datetime and make a series of datetime with your plan
# D. Search date by any hobby input.

#ORDER A
print("=="*10+"Order A"+"=="*10)
PlanString = "Hiking at weekend,Swimming on Monday,Martial arts on wednesday,Jogging on Friday,Football on sunday"
PlanList = PlanString.split(",")
display(PlanList)
HobbyInput = input("Please input your Hobby: ")
Check = [value for value in PlanList if HobbyInput.lower() in value.lower()]
if(Check != []):
    print(f"Your plan is currently having your hobby:{Check}")
else:
    print("Your plan isn't having your hobby !")

#ORDER B
print("=="*10+"Order B"+"=="*10)
CheckString = ''
CheckString = CheckString.join(Check)
print(CheckString)
NewPlan = input("Please input new plan that you want to replace: ")
dem = -1
for i in PlanList:
    if CheckString in i and Check != []:
        dem+=1
        PlanList[dem] = NewPlan
display(PlanList)

#ORDER C
print("=="*10+"Order C"+"=="*10)
DateTime = pd.DatetimeIndex(['2021-06-10','2020-07-10','2021-09-10','2021-04-11','2021-6-25'])
PlanSeries = pd.Series(PlanList, index = DateTime)
print(PlanSeries)

#ORDER D
print("=="*10+"Order D"+"=="*10)
HobbyInput = input("Input your hobby and find plan with date: ")
result = [(key.day,key.month,key.year,value) for (key,value) in PlanSeries.items() if HobbyInput.lower() in value.lower()]
print("Result of searching: ",result)
```

```python
# Question 8: Input four lists of Customer, Product, Quantity
# A. Find order information of a specific customer
# B. Find customer information who bought Pork or Vegetable
# C. Find customer information who bought Pork over 2kg
CustomerList = ["John", "John", "Marry", "Marry", "Marry"]
ProductList = ["Beer", "Pork", "Milk", "Vegetable", "Pork"]
QuantityList = ["2 Bottles", "1 kg", "5 boxes", "2 bunches", "3 kg"]

print("==" * 5 + "A. Find order information of a specific customer" + "==" * 5)
SearchStr1 = "John"
FoundIdxList = np.where(np.array(CustomerList) == SearchStr1)[0]
print(f"Order Information of Customer {SearchStr1}")
FoundProductList = [ProductList[index] for index in FoundIdxList]
FoundQuantityList = [QuantityList[index] for index in FoundIdxList]
print(f"\t ProductList = {FoundProductList}")
print(f"\t Quantity = {FoundQuantityList}")

print("==" * 5 + "B. Find customer information who bought Pork or Vegetable" + "==" * 5)
Condition1 = np.array(ProductList) == "Pork"
Condition2 = np.array(ProductList) == "Vegetable"  
print(f"Test : Condition 1 : {Condition1} and Condition 2 : {Condition2}")
FoundCustomerList = np.array(CustomerList)[Condition1 | Condition2]
FoundCustomerList = list(set(FoundCustomerList))
print(f"Found Customer : {FoundCustomerList} ")

print("==" * 5 + "C. Find customer information who bought Pork over 2kg" + "==" * 5)
NumQuantityList = [int(x.split()[0]) for x in QuantityList]
UnitList = [x.split()[1].lower() for x in QuantityList]
Condition3 = np.where(np.array(NumQuantityList) >= 2, True, False)
Condition4 = np.where(np.array(UnitList) == "kg", True, False)
print(f"Test : Condition 1 : {Condition1} and Condition 4 : {Condition4}")
FoundCustomerList1 = np.array(CustomerList)[Condition1]
FoundCustomerList2 = np.array(CustomerList)[Condition3 & Condition4]
FoundCustomerList = list(set(FoundCustomerList1) & set(FoundCustomerList2))
print(f"Found Customer : {FoundCustomerList} ")
```

```python
# Question 9: Input four lists of ItemName, Color, Weight, Price 
# A. Find product information (ItemName, Color, Weight, Price) with specific ItemName 
# B. Find Product information which has the color in list Red, Green, Blue
# C. Find product information with maximum Price 
# D. Find product information with Price over 1000 and have type Pork in ItemName
ItemNameList = ["Car", "Bicycle", "Motorbike", "Plane", "Canoe", "Rocket", "Pork","Car","Pork"]
ColorList = ["Purple", "Blue", "Green", "Blue", "Pink", "White", "Red", "Red","Red"]
WeightList = ["200 Kg", "8 Kg", "70 Kg", "450 Tons", "2 Tons", "1000 Tons", "0.5 Kg","400 kg","1 ton"]
PriceList = ["4500 $", "150.99 $", "1500.69 $", "2000000 $", "50000 $", "4000000 $", "3.99 $","3000 $", "1500 $"]

print("=="*10+"Order A: Find product information (ItemName, Color, Weight, Price) with specific ItemName "+"=="*10)
ItemNameInput = input("Please input an item name and return info: ").title()
FoundOrderA = np.where(np.array(ItemNameList) == ItemNameInput)[0]
FoundColorListA = [ColorList[index] for index in FoundOrderA]
FoundWeightListA = [WeightList[index] for index in FoundOrderA]
FoundPriceListA = [PriceList[index] for index in FoundOrderA]
if(FoundOrderA.size > 0):
    print(f"Order info of Item names {ItemNameInput}")
    print(f"\t Color: {FoundColorListA}")
    print(f"\t Weight: {FoundWeightListA}")
    print(f"\t Price: {FoundPriceListA}")
else:
    print(f"Can't find any info about {ItemNameInput}")

print("=="*10+"Order B: Find Product information which has the color in list Red, Green, Blue "+"=="*10)
ConditionB1 = np.array(ColorList) == "Blue"
ConditionB2 = np.array(ColorList) == "Green"
ConditionB3 = np.array(ColorList) == "Red"
FoundOrderB = np.array(ItemNameList)[ConditionB1 | ConditionB2 | ConditionB3]
FoundOrderB = list(set(FoundOrderB))
if(FoundOrderB != []):
    print(f"\t Found Items: {FoundOrderB}")
else:
    print("Can't find any items relating to blue,green,red !")
    
print("=="*10+"Order C: Find product information with maximum Price "+"=="*10)
PriceNumber = [float(x.split()[0]) for x in PriceList]
MaxPrice = max(PriceNumber)
ConditionC1 = np.where(np.array(PriceNumber) == MaxPrice, True, False)
FoundOrderC = np.array(ItemNameList)[ConditionC1]
FoundOrderC = list(set(FoundOrderC))
FoundOrderCstr = FoundOrderCstr.join(FoundOrderC)
print(f"\t {FoundOrderCstr} has the highest price: {MaxPrice} $")

print("=="*10+"Order D: Find product information with Price over 1000 and have type Pork in ItemName "+"=="*10)
ConditionD1 = np.where(np.array(PriceNumber) > 1000, True, False)
ConditionD2 = np.array(ItemNameList) == "Pork"
FoundOrderDColor = np.array(ColorList)[ConditionD1 & ConditionD2]
FoundOrderDWeight = np.array(WeightList)[ConditionD1 & ConditionD2]
FoundOrderDPrice = np.array(PriceList)[ConditionD1 & ConditionD2]
print(f"Name: Pork")
print(f"\t Color: {FoundOrderDColor}")
print(f"\t Weight: {FoundOrderDWeight}")
print(f"\t Price: {FoundOrderDPrice}")
```

```python
# Question 10: Input four lists of Customer, Product, Quantity
# A. Create a DataFrame from three lists above 
# B. Seperate column QuantityList to Quantity and Unit
# C. Find customer information who bought Pork over 2kg 
CustomerList = ["John", "John", "Marry", "Marry", "Marry"]
ProductList = ["Beer", "Pork", "Milk", "Vegetable", "Pork"]
QuantityList = ["2 Bottles", "1 kg", "5 boxes", "2 bunches", "3 kg"]

print("==" * 5 + "A. Create a DataFrame from three lists above" + "==" * 5)
d = {'Customer' : CustomerList, 'Product' : ProductList, 'QuantityList' : QuantityList}
df = pd.DataFrame(d)
display(df)
print(f"Shape of DataFrame : {df.shape}, with rows : {df.shape[0]}, and columns : {df.shape[1]}")

print("==" * 5 + "B. Seperate column QuantityList to Quantity and Unit" + "==" * 5)
df[["Quantity","Unit"]] = df["QuantityList"].str.split(" ",expand=True)
df["Unit"] = df["Unit"].str.lower()
df["Quantity"] = df["Quantity"].apply(pd.to_numeric)
display(df)

print("==" * 5 + "C. Find customer information who bought Pork over 2kg" + "==" * 5)
Condition1 = df.Product.str.lower() == "pork"
Condition2 = df.Quantity > 2
Condition3 = df.Unit == "kg"
CustomerDF = df[Condition1 & Condition2 & Condition3]
display(CustomerDF)
print(f'Found : {list(CustomerDF["Customer"])}')
```
