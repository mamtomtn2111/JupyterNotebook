```python
# Question 1: Write a function to input a number N (N >500) and two more numbers (mean,std) 
# and do the following things:
# A. Generate an array with N numbers with two decimals following Gauss Distribution(mean, std) 
# B. Print the statstics information : min, max, mean, standard deviation, Q1, Q2, Q3, IQR
# C. If we have a rule : a outlier of list is a number are not in the range (Q1 - 1.5*IQR, Q3 + 1.5*IQR). 
#    print all the outlier number and its position. 

def printStatsOnArray(N = 500, mu = 0, sigma = 0.1):
    np.random.seed(12)
    NumArray =  np.random.normal(mu, sigma, N)
    NumArray =  np.round(NumArray, 2)
    print(f"The generated array about {len(NumArray)} element under Gauss Distribution is : \n{NumArray[:30]}")

    StatsResult = NumArray.min(), NumArray.max(), \
                  NumArray.mean(), NumArray.std(),\
                  np.quantile(NumArray, 0.25), np.quantile(NumArray, 0.5), np.quantile(NumArray, 0.75), \
                  np.quantile(NumArray, 0.75) - np.quantile(NumArray, 0.25)  
    StatsResult = np.round(StatsResult, 5)

    NormalRange = [StatsResult[4] - 1.5*StatsResult[-1], StatsResult[6] + 1.5*StatsResult[-1]]
    OutlierList = [(idx,val) for idx, val in enumerate(NumArray) if val < NormalRange[0] or val > NormalRange[1]]
    print(f"The statistics of Array is : \n[min, max, mean, standard deviation, Q1, Q2, Q3, IQR] \n{StatsResult}")
    print(f"Range of normal value is {NormalRange}")
    print(f"The Outlier list is \n{OutlierList}" if len(OutlierList) > 0 else "There is no outlier")
    return
    
def main():
    N = 1000
    mu, sigma = 0, 0.1 # mean and standard deviation
    printStatsOnArray(N, mu, sigma)

if __name__ == '__main__':
    main()
```

```python
# Question 2: Write a function to input a number N and start, stop  
# and do the following things:
# A. Random a N elements of floating number and round to 2 decimals
# B. Seperate the generated list into the Integer and Decimal Part List 
# C. Make the three test cases : with start and stop, only start, and no start or stop

def FindMixMax(N, start = None, stop = None):
    if(not start and not stop): 
        start = 0
        stop = 10
    if(not start): start = 0
    if(not stop) : stop = 100
        
    np.random.seed(12)
    RandList = [round(random.uniform(start, stop), 2) for i in range(N)]
    IntegerList = [int(x) for x in RandList]
    DecimalPart = [round(u - v,2) for u,v in zip(RandList, IntegerList)]
    return RandList, IntegerList, DecimalPart

# Test 1
N = 10
start = 1
stop = 1000
print("Test 1")
NumList, IList, DList = FindMixMax(N, start, stop)
print(f"The Integer and Decimal Part of List \n{NumList} is \n Integer Part : {IList} \n Decimal Part:{DList}")

#Test 2
print("Test 2")
NumList, IList, DList = FindMixMax(N, start)
print(f"The Integer and Decimal Part of List \n{NumList} is \n Integer Part : {IList} \n Decimal Part:{DList}")

#Test 3
print("Test 3")
NumList, IList, DList = FindMixMax(N)
print(f"The Integer and Decimal Part of List \n{NumList} is \n Integer Part : {IList} \n Decimal Part:{DList}")
```

```python
# Question 4: Write a function with no input  
# and do the following things:
# A. Ask user to input continuously a name of user, a name of product and a comment of that user about product
# until they input username is blank
# B. Each input of username, productname, comment add to a row of dataframe 
# with three columns (CustomerName, ProductName, and Comment) 
# C. Return that dataframe and display it

def InputCustomer():
    UserName = []
    ProductName = []
    Comment = []
    print("Please input continuously your name, a name of product and a comment about it")
    i = 1
    
    while(True):
        print(f"=====User[{i}]=====")
        InputName = input("User name?: ").title()
        if(len(InputName.strip()) == 0):
            print("You have just stopped the input procedure")
            break
        InputProduct = input("Product name?: ").strip()
        InputComment = input("Comment about it?: ").strip()
        i+=1
        UserName.append(InputName)
        ProductName.append(InputProduct)
        Comment.append(InputComment)
    return UserName,ProductName,Comment

def DfCreater(Name,Product,Comment):
    d = {'User Name':Name,'Product name':Product,'Comment':Comment}
    df = pd.DataFrame(d)
    display(df)
    
def main():
    Name, Product, Comment = InputCustomer()
    DfCreater(Name,Product,Comment)
    
if __name__ == "__main__":
    main()
```

```python
# Special Remember
def myfamily(*kids):
    print("\tThe oldest child is " + kids[0])
    print("\tThe youngest child is " + kids[-1])

def calculation(*numberparms):
  sum = 0 
  for num in numberparms:
    sum = sum + num
  mean = sum/len(numberparms)
  print("\tSum = {} and Mean = {}".format(sum,mean))
  return

print("myfamily : ")
myfamily("Tom", "Teo", "Ti", "Bin", "Bo")
print("calculation")
calculation(1, 2, 3, 4, 5)
```

```python
# Special Remember
def statistics(NumList):
  import numpy as np
  sum = np.sum(NumList)
  mean = np.mean(NumList)
  count = len(NumList)
  min = np.min(NumList)
  max = np.max(NumList)
  return sum, mean, count, min, max

sum, mean, count, min, max = statistics([1,2,3,4,5])
result = statistics([1,2,3,4,5])
print(sum, mean, count, min, max)
print(statistics([1,2,3,4,5]))
print(result, type(result))
for r in result:
  print(r)
```
