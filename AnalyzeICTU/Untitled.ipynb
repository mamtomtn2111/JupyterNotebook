```python
import pandas as pd
import numpy as np
```

```python
Data = pd.read_excel("CNTTICTU.xlsx")
Data.head(10)
```

```python
K18Data = Data[Data['Lớp'].str.contains('K18')]
K18Data.head(10)
```

```python
K18CountSV = K18Data["Lớp"].value_counts().rename_axis('Lớp').reset_index(name='Số học sinh')
print("Tổng số có lớp CNTT:"+str(K18CountSV["Lớp"].count()))
display(K18CountSV)
```
