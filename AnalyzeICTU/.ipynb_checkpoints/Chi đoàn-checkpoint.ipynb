```python
import pandas as pd
```

```python
form = pd.read_excel("C:\\Users\\mamto\\OneDrive\\Máy tính\\Đại hội\\google2.xlsx")
VuLuc = pd.read_excel("C:\\Users\\mamto\\OneDrive\\Máy tính\\Đại hội\\VuLuc.xlsx")
newdf = VuLuc.merge(form, on = ["Tên Sinh Viên"], how = "left")
print(newdf.keys())
newdf.drop(newdf.filter(regex="Unname"),axis=1, inplace=True)
newdf = newdf.fillna("Chưa điền")
display(newdf)
newdf.to_excel("C:\\Users\\mamto\\OneDrive\\Máy tính\\Đại hội\\Thông tin điền form.xlsx")
```

```python
JoinState = newdf["Bạn có thể tham gia không ?"].value_counts().rename_axis('Trạng thái tham gia').reset_index(name='counts')
display(JoinState)
my_explode=(0,0,0.1)
plot = JoinState.plot.pie(y = "counts", labels=JoinState['Trạng thái tham gia'], legend = False, explode = my_explode,autopct='%1.1f%%')
```

```python
Allfalcuty = pd.read_excel("D:\\Work\\Đoàn thanh niên\\THỐNG KÊ KHẢO SÁT KHẢO SÁT Ý KIẾN SINH VIÊN VỀ CÔNG TÁC ĐÁNH GIÁ HỌC TRỰC TUYẾN.xlsx")
Allfalcuty.head(5)
```

```python
ITfalcuty = Allfalcuty[Allfalcuty["Khoa"] == "Khoa Công nghệ thông tin"]
ITfalcuty.shape
```

```python
Class = ITfalcuty["Lớp"].value_counts().rename_axis('Lớp').reset_index(name='Số sinh viên')
Class = Class.sort_values(by="Lớp",ascending = True,inplace = False)
Class =  Class[~Class["Lớp"].str.contains("K16")].reset_index(drop=True)
Class
#Class.to_excel("D:\\Work\\Đoàn thanh niên\\Triển khai nghiên cứu khoa học.xlsx")
```
